```python
df["sex"].value_counts()
```

```python
sb.countplot(x="target", data=df)

plt.title("Heart Disease Distribution")
plt.grid(True)
plt.savefig("images/number of patients_distribution.png",dpi = 300,bbox_inches = 'tight')
```

```python
plt.figure(figsize=(12,10))
sb.histplot(df['age'], bins = 60)
plt.title("Age of patients")
plt.xlabel("Age")
plt.ylabel("no.of Patients")
plt.grid(True)
plt.savefig("images/age_distribution.png",dpi = 300,bbox_inches = 'tight')
```

```python
sb.countplot(
    x = "sex",
    hue = "target",
    data = df
)      

plt.title("Gender vs patients")
plt.xlabel("Gender")
plt.ylabel("Number")
plt.grid(True)
plt.savefig("images/Gender Vs Patients.png",dpi = 300,bbox_inches = 'tight')
```

```python
sb.boxplot(
    x="target",
    y="chol",
    data = df
)

plt.grid(True)
plt.title("Cholestrol Analysis")
plt.savefig("images/cholestrol_Analysis.png",dpi = 300,bbox_inches = 'tight')
```

```python
sb.lineplot(
    x="target",
    y="fbs",
    data=df
)

plt.grid(True)
plt.savefig("images/Blood_Sugar distribution.png",dpi = 300,bbox_inches = 'tight')
```

```python
sb.boxplot(
    x="target",
    y="trestbps",
    data = df
)

plt.grid(True)
plt.savefig("images/Blood pressure_distribution.png",dpi = 300,bbox_inches = 'tight')
```

```python
corr_matrix = df.corr(numeric_only = True)
corr_matrix
```

```python
plt.figure(figsize=(12,10))

sb.heatmap(
    corr_matrix,
    annot = True,   
)

plt.title("Correlation Heatmap")
plt.xlabel("Factors")
plt.savefig("images/HeatMap_distribution.png",dpi = 300,bbox_inches = 'tight')
```
